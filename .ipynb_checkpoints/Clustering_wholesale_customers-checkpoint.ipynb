## cluster Miniproject
Multiple Dunn Index measures resembling different k used for K-Means clustering your data.
An output plot of the elbow curve.
The best k chosen based on the elbow curve plot.
Output predicted clusters for the first 10 data samples.

```python
## load libraries
import pandas as pd
import seaborn as sbn 
from matplotlib import pyplot as plt
import numpy as np

from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans

from sklearn.metrics import silhouette_score
from sklearn.metrics import davies_bouldin_score
```

```python
## Read the dataset into a Pandas Data Frame!
df=pd.read_csv("Wholesale_customers_data.csv")

# drop missing data
df_drop= df.dropna()
df
```

```python
# Feature Selection
## Either exclude the Channel and Region from the considered features 
df_selected = df_drop.drop(['Channel', 'Region'], axis=1)
df_selected
```

```python
## Exploratory data plots:

for col in df_selected.columns:
    hist = df_selected[col].hist(bins=10)
    print("Plotting for column {}".format(col))
    plt.show()
```

```python
# A correlation matrix heat map.

plt.figure(figsize=(16, 6))
heatmap = sbn.heatmap(df_selected.corr(), annot=True)
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12)
```

```python
## Preprocessing
# Scaling
scaler = MinMaxScaler(feature_range=(0,1))
df_scaled = pd.DataFrame(scaler.fit_transform(df_selected.values), columns=df_selected.columns, index=df_selected.index)
df_scaled
```

```python
### Elbow plot ######

### inertia values for different K:
distortions = []
K = range(1,20)
for k in K:
   kmeanModel = KMeans(n_clusters=k)
   kmeanModel.fit(df_scaled)
   distortions.append(kmeanModel.inertia_)

## Plot distortions vs Values of K
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
## get the inertia values from each iteration from kmeans 
sse = []
for i in range(3,50):
    kmeans = KMeans(n_clusters=i , max_iter=300, tol=0.01)
    kmeans.fit(df_scaled)  # <- fit here.....
    sse.append(kmeans.inertia_)
    
plt.plot(range(3,50),sse)
plt.xlabel("Number of Custers")
plt.ylabel("Inertia")
plt.show()
```

```python
## Clustering Using K-Means: Use a K-Means instance to cluster the data.
## given the elbolw plot: I choose a k of 5. 

kmeans = KMeans(n_clusters=5, tol = 0.01)
kmeans.fit(df_scaled)
#predict the clusters of data.
clusters = kmeans.predict(df_scaled)
print(clusters)
kmeans.inertia_
```

```python
## Labels for first 10 samples
labels=kmeans.labels_
labels[1:11]
```

```python
## Scoring

SS= silhouette_score(df_scaled, clusters)
DBS = davies_bouldin_score(df_scaled, clusters)
print("SS: ",SS)
print("DBS: ", DBS)
```

## conclusion
The elbow plot we consider 5 to be the optimum number of clusters. As the silhouette score is between 0.25 - 0.5: It is a fair clustering. Hierarchical clustering is also tried next.

##  Hierarchical Clustering.

```python
from scipy.cluster.hierarchy import linkage, dendrogram

# Calculate the linkage: mergings
mergings = linkage(df_scaled,method='complete')
mergings
```

```python
# Plot the dendrogram
dendrogram(mergings,
           leaf_rotation=90,
           leaf_font_size=6,
)
plt.show()
```
